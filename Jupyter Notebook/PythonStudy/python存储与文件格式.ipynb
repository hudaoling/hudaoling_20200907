# 逐块读取文本文件

```python
#nrows=n，代表取表格中前n行的值
df=pd.read_csv('F:/Jupyter Files/testdata/T_Area.csv',nrows=10)
```

```python
df
```

```python
#chunksize=1000指定逐块读取文件
chunker=pd.read_csv('F:/Jupyter Files/testdata/T_Area.csv',chunksize=1000)
chunker
```

```python
chunker=pd.read_csv('F:/Jupyter Files/testdata/T_Area.csv',chunksize=1000)
tot=Series([])#空的数组
for piece in chunker:
    tot=tot.add(piece['Id'].value_counts(),fill_value=0) #在trunker行里面的计数
tot=tot.sort_index(ascending=False)
```

```python
tot
```
