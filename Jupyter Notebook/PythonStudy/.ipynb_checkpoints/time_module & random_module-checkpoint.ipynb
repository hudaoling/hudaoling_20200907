# time

```python
import time
```

***time的三种格式***
+ 时间戳:time.time()
+ 格式化字符串日期 "2016-08020 14:08:01"
+ 元组格式日期struct_time:   time.localtime()

***时间戳转化成元组struct_time(tuple)，***  
***gmtime:结果为UTC时区（格林威治天文时间）***  
***localtime:结果为UTC+8时区（北京时间）***  
+ print(time.localtime())#UTC+8北京时间
+ print(time.gmtime())#UTC时区，不输入默认传当前时间戳
+ print(time.gmtime(10000000))#也可以自定义输入的时间戳

```python
time.sleep(1) #等待1秒
```

```python
time.time() #时间戳
```

```python
time.localtime() #元组格式日期
```

```python
time.localtime(time.time()) #时间戳转换成元组格式
```

```python
time.mktime(time.localtime())#将元组格式转换成时间戳格式
```

```python
time.localtime(222222222)
```

```python
time.asctime(time.localtime())#元组转换成系统默认的日期字符格式
```

```python
time.ctime(time.time())#时间戳转成默认的日期字符格式
```

```python
x=time.strftime("%Y-%m-%d %H:%M:%S",time.localtime()) #将元组转换成格式化字符串日期
x
```

```python
#时间戳不能直接转换为格式化字符串，需先用time.localtime()转换成元组
time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(2222222222))
```

```python
time.strptime(x,"%Y-%m-%d %H:%M:%S")#将字符串格式转换成元组格式日期
```

```python
z=time.localtime(22222222222)
z
```

```python
#取出日期的年月日，时分秒,只适合元组格式取值
z.tm_year,z.tm_mon,z.tm_mday,z.tm_hour,z.tm_min,z.tm_sec
```

```python
#随机生成验证码的例子
checkcode=''
for i in range(5):
    current=random.randrange(0,5)
    if current == i:
        tmp=chr(random.randint(60,90))
    else:
        tmp=random.randint(0,9)
    checkcode+=str(tmp)
print(checkcode)
```
