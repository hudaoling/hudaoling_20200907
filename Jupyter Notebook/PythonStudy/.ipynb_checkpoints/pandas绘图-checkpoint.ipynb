# pandas绘图

```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as  np
from numpy import reshape
from numpy import cumsum
import matplotlib.pyplot as plt
```

```python
 plt.close('all')
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts=ts.cumsum()
```

```python
ts.plot()
```

```python
#columns设置标记，相当于图例
df = pd.DataFrame(np.random.randn(1000, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'], 
                  index=ts.index)
                 # index=np.arange(0, 10000, 10))   #索引可以支持日期也可以支持数字
```

```python
df.plot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3.head()
```

```python
df3['A']=pd.Series(list(range(len(df))))
df3.head()
```

```python
#指定X,Y轴使用哪个字段（列或叫维度）
df3.plot(x='A',y='B')
```

```python
plt.figure()
```

```python
#创建条形图,iloc[5]获取前4行
df.iloc[5].plot(kind='bar')
```

# 直方图df.plot.hist

```python
df4 = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000),'c': np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
df4.head()
```

```python
plt.figure()
```

```python
df4.plot.hist(alpha=0.5)#alpha=0.5代表颜色深浅，从浅0.1-1深的取值
```

```python
#堆叠直方图stacked=True。料箱尺寸可以使用bins关键字进行更改。
df4.plot.hist(stacked=True, bins=20)
```

#箱线图 df.plot.box
+ 箱线图可以得出呼叫Series.plot.box()和DataFrame.plot.box()，或DataFrame.boxplot()到每一列内可视化值的分布。

```python
 df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.head()
```

```python
df.plot.box()
```

```python
color = {'boxes': 'DarkGreen', 'whiskers': 'DarkOrange','medians': 'DarkBlue', 'caps': 'Gray'}
```

```python
#指定箱线图的颜色
df.plot.box(color=color, sym='r+')
```

```python
#您可以传递matplotlib支持的其他关键字boxplot。例如，可以使用vert=False和positions关键字绘制水平和自定义位置的箱线图 。
df.plot.box(vert=False, positions=[1, 4, 5, 6, 8])
```

# 面积图df.plot.area
+ 您可以使用Series.plot.area()和创建面积图DataFrame.plot.area()。区域图默认情况下是堆叠的。要生成堆积面积图，每列必须全部为正值或全部为负值。
+ 当输入数据包含NaN时，它将自动用0填充。如果要删除或填充不同的值，请在调用plot之前使用dataframe.dropna()或。dataframe.fillna()

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.area()
```

```python
#要生成未堆积的图，请通过stacked=False。除非另有说明，否则Alpha值将设置为0.5：
df.plot.area(stacked=False);
```

# 散点图df.plot.scatter()
+ 使用该DataFrame.plot.scatter()方法可以绘制散点图。散点图要求x和y轴为数字列。这些可以通过x和y关键字指定。

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df.plot.scatter(x='a', y='b');
```

```python
ax = df.plot.scatter(x='a', y='b', color='DarkBlue', label='Group 1')
```

```python
 df.plot.scatter(x='c', y='d', color='DarkGreen', label='Group 2', ax=ax)
```

```python
 df.plot.scatter(x='a', y='b', c='c', s=50);
```

```python
 df.plot.scatter(x='a', y='b', s=df['c'] * 200);
```

# 饼图 
+ 您可以使用DataFrame.plot.pie()或创建饼图Series.plot.pie()。如果您的数据包含any NaN，它们将自动填充为0。如果您的数据中包含任何ValueError负值，则将引发A。

```python
series = pd.Series(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], name='series')
series.plot.pie(figsize=(6, 6))
```

```python
df = pd.DataFrame(3 * np.random.rand(4, 2),index=['a', 'b', 'c', 'd'], columns=['x', 'y'])
df.head()
```

```python
#请注意，带有的饼图DataFrame要求您通过y参数或指定目标列subplots=True。当y指定，选定列的饼状图将被绘制。
#如果subplots=True指定为，则将每列的饼图绘制为子图。
df.plot.pie(subplots=True,figsize=(8, 4))
```

```python
series.plot.pie(labels=['AA', 'BB', 'CC', 'DD'], colors=['r', 'g', 'b', 'c'],autopct='%.2f', fontsize=20, figsize=(6, 6))
```

```python
 df.plot(subplots=True, figsize=(6, 6))
```
