# IPython
+ **绘图操作的GUI控制台**
+ **Web的交互式笔记本**
+ **轻量级的快速并行计算引擎**  
如何启动：cmd进入命令行操作，输入ipython即启动完成

+ ***Tab键自动完成***：命令和方法自动匹配
+ ***使用？号对象内省本***：显示该对象的一些通用信息
+ ***%run命令***：执行python文件 例如: %run test.py
+ ***中断正在执行的代码***：Ctrl+C
+ ***执行剪贴板中的代码***：%paste或%cpaste两个魔术函数；Ctrl+Shift+V常规的代码粘贴
+ ***键盘快捷键***：Ctrl+C终止代码运行；Ctrl+F或b将光标向前或向后移动一个字符；Ctrl+L清屏等（参照书P56）
+ ***异常跟踪***：脚本或语句执行异常时，默认输出调用栈跟踪，traceback。
+ ***魔术命令***：%magic;%debug;%reset;%paste;%run等
+ ***使用命令历史，搜索并重用命令历史***：Ctrl+P/Ctrl+N,Ctrl+R（参照书P61）
+ ***记录输入输出，输入输出变量***：
+ ***与操作系统交互***：%pwd返回系统当前工作目录；%env 以dict的形式返回系统变量等等（参照书P63）
+ ***交互式调试器***：
+ ***基本性能分析***：%prun和%run-p,宏观性能分析
+ ***逐行分析函数性能***：%lprun和%run-p，微观性能分析
