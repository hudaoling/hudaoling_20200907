```python
import urllib.request
import urllib.parse
import string #url汉字转码
```

```python
def load_data():
    url = "http://www.baidu.com/"
    #get请求，http请求
    #response:http相应的对象
    response = urllib.request.urlopen(url) #返回：HTTPResponse
    data = response.read() #读出来是bytes
    str_data=data.decode("utf-8") #转换成字符串
    
    #持久化数据：写入并保存到文件/数据库
    with open("baidu.html","w",encoding="utf-8") as f:
        f.write(str_data)
    print('HTTPResponse:',response,'\n')
    print('data:',data,'\n')
    print('str_data:',str_data)
#主要的python爬取的类型：str bytes

load_data()
```

```python
#讲字符串类型转换成bytes
str_name="baidu"
str_to_bytes_name=str_name.encode("utf-8") 
bytes_to_str=str_to_bytes_name.decode("utf-8")#bytes转换成str时，用dedode
print(str_to_bytes_name)
print(bytes_to_str)
```

```python
def get_method_params():

    url = "http://www.baidu.com/s?wd="
    #拼接字符串(汉字)
    #python可以接受的数据
    #https://www.baidu.com/s?wd=%E7%BE%8E%E5%A5%B3

    name = "图谱"
    final_url = url+name
    print(final_url)
    #代码发送了请求
    #网址里面包含了汉字;ascii是没有汉字的;url转译
    #将包含汉字的网址进行转译
    encode_new_url = urllib.parse.quote(final_url,safe=string.printable)
    print(encode_new_url)
    # 使用代码发送网络请求
    response = urllib.request.urlopen(encode_new_url)
    print(response)
    #读取内容
    data = response.read().decode()
    print(data)
    #保存到本地
    with open("02-encode.html","w",encoding="utf-8")as f:
        f.write(data)
    #UnicodeEncodeError: 'ascii' codec can't encode
    # characters in position 10-11: ordinal not in range(128)
    #python:是解释性语言;解析器只支持 ascii 0 - 127
    #不支持中文

get_method_params()
```

# json格式转换和读写

```python
import json
 
# Python 字典类型转换为 JSON 对象
data = {
    'no' : 1,
    'name' : 'Runoob',
    'url' : 'http://www.runoob.com'
}
 
json_str = json.dumps(data)
print ("Python 原始数据：", repr(data))
print ("JSON 对象：", json_str)
type(json_str)
```

```python
# 将 JSON 对象转换为 Python 字典
data2 = json.loads(json_str)
print ("data2['name']: ", data2['name'])
print ("data2['url']: ", data2['url'])
type(data2)
```

```python
# 写入 JSON 数据
with open('data.json', 'w') as f:
    json.dump(data, f)
```

```python
 
# 读取数据
with open('data.json', 'r') as f:
    data = json.load(f)
    print(data)
```
